## Lab 3 - Bài 1

```python
DaySo = [57,45,11,23,67,32,89,90,12,24]
#sắp xếp danh sách
print("Danh sách ban đầu: ",DaySo)
DaySo.sort()
print("Danh sách đã sắp xếp: ",DaySo)
#tìm phần tử nhỏ nhất
print("Phần tử nhỏ nhất: ",DaySo[0])
#tính tổng & đếm các số chia hết cho 3
Tong = 0
Dem = 0 
for i in DaySo:
    if i % 3 == 0:
        Tong += i
        Dem += 1
print("Tổng các số chia hết cho 3: ",Tong)
print("Đếm các số chia hết cho 3: ",Dem)
#tính kết quả trung bình
if Dem > 0:
    TrungBinh = Tong / Dem
    print("Kết quả trung bình các số chia hết cho 3: ",TrungBinh)
else:
    print("Không có số nào chia hết cho 3")

#nâng cao ( sắp xếp giảm dần, tìm phần tử nhỏ nhất)
DaySo.sort(reverse=True)
print("Danh sách sắp xếp giảm dần: ",DaySo)
print("Phần tử nhỏ nhất là phần tử thứ: ",DaySo[-1])


```

```python
#nhập dữ liệu hs Cú pháp: <<tên biến>> = {<<khóa 1: giá trị 1>>, <<khóa 2: giá trị 2>>, …}
HocSinh = {
    "Họ tên": "Nguyễn Văn A", 
    "Tuổi": 19, 
    "Lớp": "12A1", 
    "Điểm toán": 8.5, 
    "Điểm lý": 7.0, 
    "Điểm hóa": 6.5
}

diemtrungbinh = (HocSinh["Điểm toán"] + HocSinh["Điểm lý"] + HocSinh["Điểm hóa"]) / 3
if diemtrungbinh >= 9:
    xeploai = "Xuat sắc"
elif 8 <= diemtrungbinh < 9:
    xeploai = "Giỏi"
elif 7 <= diemtrungbinh < 8:
    xeploai = "Khá"
elif 5 <= diemtrungbinh < 7:
    xeploai = "Trung bình"
else:
    xeploai = "Yếu"

print("Họ tên:", HocSinh["Họ tên"])
print("Điểm trung bình:", round(diemtrungbinh, 2))
print("Xếp loại:", xeploai)
```

Lab 3 - Bài thêm

```python
# Danh sách sinh viên
DanhSachSV = []

# Nhập số lượng sinh viên
n = int(input("Nhập số lượng sinh viên: "))

for i in range(n):
    print(f"\n--- Nhập thông tin sinh viên {i+1} ---")
    hoten = input("Họ tên: ")
    diemtrungbinh = float(input("Điểm trung bình: "))

    # Tạo dictionary cho sinh viên
    sv = {
        "Họ tên": hoten,
        "Điểm trung bình": diemtrungbinh
    }

    # Thêm vào danh sách
    DanhSachSV.append(sv)

print("\n=== KẾT QUẢ ===")
for sv in DanhSachSV:
    diemtrungbinh = sv["Điểm trung bình"]

    # Xếp loại
    if diemtrungbinh >= 9:
        xeploai = "Xuất sắc"
    elif 8 <= diemtrungbinh < 9:
        xeploai = "Giỏi"
    elif 7 <= diemtrungbinh < 8:
        xeploai = "Khá"
    elif 5 <= diemtrungbinh < 7:
        xeploai = "Trung bình"
    else:
        xeploai = "Yếu"
    
    # Xuất kết quả
    print("\nHọ tên:", sv["Họ tên"])
    print("Điểm trung bình:", round(diemtrungbinh, 2))
    print("Xếp loại:", xeploai)
```

## Lab 3 - Bài 3

```python
# Nhập số nguyên từ bàn phím
n = int(input("Nhập một số nguyên: "))

# Giả sử ban đầu là số nguyên tố
ok = True
i = 2

# Kiểm tra từ 2 đến n-1
while i < n:
    if n % i == 0:
        ok = False
        break
    i += 1

# In kết quả
if n > 1 and ok:
    print(n, "là số nguyên tố")
else:
    print(n, "không phải là số nguyên tố")
```

## Lab 3 - Bài 5: bài tập làm thêm

```python
import math
import datetime

# Lưu lịch sử phép tính
lich_su = []

def phep_tinh_co_ban():
    a = float(input("Nhập số thứ nhất: "))
    b = float(input("Nhập số thứ hai: "))
    print("1. Cộng (+)")
    print("2. Trừ (-)")
    print("3. Nhân (*)")
    print("4. Chia (/)")
    chon = input("Chọn phép tính: ")
    if chon == "1":
        kq = a + b
        phep = f"{a} + {b} = {kq}"
    elif chon == "2":
        kq = a - b
        phep = f"{a} - {b} = {kq}"
    elif chon == "3":
        kq = a * b
        phep = f"{a} * {b} = {kq}"
    elif chon == "4":
        if b != 0:
            kq = a / b
            phep = f"{a} / {b} = {kq}"
        else:
            phep = "Không thể chia cho 0!"
    print(phep)
    lich_su.append(phep)

def luy_thua():
    x = float(input("Nhập cơ số x: "))
    y = float(input("Nhập số mũ y: "))
    kq = x ** y
    phep = f"{x}^{y} = {kq}"
    print(phep)
    lich_su.append(phep)

def can_bac_hai():
    x = float(input("Nhập số x: "))
    if x >= 0:
        kq = math.sqrt(x)
        phep = f"√{x} = {kq}"
    else:
        phep = "Không thể tính căn bậc hai của số âm!"
    print(phep)
    lich_su.append(phep)

def ham_luong_giac():
    x = float(input("Nhập góc (độ): "))
    rad = math.radians(x)
    phep = f"sin({x})={math.sin(rad):.4f}, cos({x})={math.cos(rad):.4f}, tan({x})={math.tan(rad):.4f}"
    print(phep)
    lich_su.append(phep)

def logarit():
    x = float(input("Nhập số x: "))
    if x <= 0:
        phep = "Logarit không xác định với x <= 0!"
    else:
        print("1. log cơ số 10")
        print("2. ln (log_e)")
        print("3. log cơ số tùy chọn")
        chon = input("Chọn loại log: ")
        if chon == "1":
            phep = f"log10({x}) = {math.log10(x)}"
        elif chon == "2":
            phep = f"ln({x}) = {math.log(x)}"
        elif chon == "3":
            base = float(input("Nhập cơ số: "))
            phep = f"log_{base}({x}) = {math.log(x, base)}"
        else:
            phep = "Chọn sai!"
    print(phep)
    lich_su.append(phep)

def giai_pt_bac_1():
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    if a == 0:
        if b == 0:
            phep = "PT vô số nghiệm"
        else:
            phep = "PT vô nghiệm"
    else:
        x = -b / a
        phep = f"PT có nghiệm x = {x}"
    print(phep)
    lich_su.append(phep)

def giai_pt_bac_2():
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    c = float(input("Nhập c: "))
    if a == 0:
        phep = "Đây không phải PT bậc 2!"
    else:
        delta = b**2 - 4*a*c
        if delta < 0:
            phep = "PT vô nghiệm"
        elif delta == 0:
            x = -b / (2*a)
            phep = f"PT có nghiệm kép x = {x}"
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            phep = f"PT có 2 nghiệm: x1 = {x1}, x2 = {x2}"
    print(phep)
    lich_su.append(phep)

def xem_lich_su():
    if len(lich_su) == 0:
        print("Chưa có phép tính nào.")
    else:
        print("=== Lịch sử tính toán ===")
        for p in lich_su:
            print(p)

def thoi_gian():
    now = datetime.datetime.now()
    print("Thời gian hiện tại:", now.strftime("%H:%M:%S %d-%m-%Y"))

# Menu chính
while True:
    print("\n=== MÁY TÍNH BỎ TÚI ===")
    print("1. Phép tính cơ bản")
    print("2. Lũy thừa")
    print("3. Căn bậc hai")
    print("4. Hàm lượng giác")
    print("5. Logarit")
    print("6. Giải PT bậc 1")
    print("7. Giải PT bậc 2")
    print("8. Xem lịch sử")
    print("9. Xem thời gian")
    print("0. Thoát")

    chon = input("Chọn chức năng: ")

    if chon == "1":
        phep_tinh_co_ban()
    elif chon == "2":
        luy_thua()
    elif chon == "3":
        can_bac_hai()
    elif chon == "4":
        ham_luong_giac()
    elif chon == "5":
        logarit()
    elif chon == "6":
        giai_pt_bac_1()
    elif chon == "7":
        giai_pt_bac_2()
    elif chon == "8":
        xem_lich_su()
    elif chon == "9":
        thoi_gian()
    elif chon == "0":
        print("Thoát chương trình!")
        break
    else:
        print("Chọn sai, vui lòng chọn lại!")
```
